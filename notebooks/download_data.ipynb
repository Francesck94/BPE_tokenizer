```python
wikipedia_url="https://en.wikipedia.org/wiki/Machine_learning"
```

```python
print(wikipedia_url)
```

```python
import requests
from bs4 import BeautifulSoup

headers = {"User-Agent": "BPE-Tokenizer-Project/1.0 (educational use)"}
response = requests.get(wikipedia_url, headers=headers)
response.raise_for_status()

soup = BeautifulSoup(response.text, "html.parser")

# Remove script, style, and reference elements
for tag in soup(["script", "style", "sup", "table"]):
    tag.decompose()

# Extract text from all paragraphs in the article body
content_div = soup.find("div", class_="mw-parser-output")
paragraphs = content_div.find_all("p")
text = "\n".join(p.get_text(strip=True) for p in paragraphs if p.get_text(strip=True))

print(f"Downloaded {len(text)} characters")
print("---")
print(text[:2000])
```

```python
with open("../example_data/wikipedia_page.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
